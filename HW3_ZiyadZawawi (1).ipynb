```python
# Importing libraries needed
import numpy as np
import pandas as pd
import uproot
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Opening files using uproot 
file = uproot.open("df_1.root")
file2 = uproot.open("df_2.root")
file3 = uproot.open("df_3.root")
file4 = uproot.open("df_4.root")
file5 = uproot.open("df_5.root")
file6 = uproot.open("df_6.root")

# Looking at files keys to open the DataFrame
file.keys()
tree = file["pandas_tree218"]

#tree.pandas.df()


tree2 = file2["pandas_tree51358"]
tree3 = file3["pandas_tree53278"]
tree4 = file4["pandas_tree52558"]
tree5 = file5["pandas_tree53038"]
tree6 = file6["pandas_tree54718"]


tree = tree.pandas.df()
tree2 = tree2.pandas.df()
tree3 = tree3.pandas.df()
tree4 = tree4.pandas.df()
tree5 = tree5.pandas.df()
tree6 = tree6.pandas.df()

# Making wholetree Dataframe which saves all the trees dataframes together in one
trees = [tree, tree2, tree3, tree4, tree5, tree6]
wholetree = pd.concat(trees, keys=["tree1", "tree2", "tree3", "tree4", "tree5", "tree6"])
# Printing the wholetree 
wholetree
```

$L_x, L_y ,$ and $L_z$ are the decay lengths in the x,y, and z directions they are considered to be 0 when IsLLP is given to be false  

IsLLP tells us if the jet has longed lived particle    
JetPt tells us the jet traversal momentum  
JetPhi tells the physicists the angle at which the particle's transversal momentum scattered   
JetEta tells us the angle of a particle relative to the beam axis 

By plotting eta and phi against each other, scientists are able to determine the distance between two particles, which can be applied to other calculations

```python
# plotting histograms of JetPT, JetEta, JetPhi, and IsLLP
JetPt = tree.array("JetPt")
plt.hist(JetPt, bins =25)
plt.title("Histogram of JetPt")
plt.ylabel("Counts")
plt.xlim([0 , 500])
plt.show()

JetEta = tree.array("JetEta")
plt.hist(JetEta, bins =25)
plt.title("Histogram of JetEta")
plt.ylabel("Counts")
plt.show()

JetPhi = tree.array("JetPhi")
plt.hist(JetPhi, bins =25)
plt.title("Histogram of JetPhi")
plt.ylabel("Counts")
plt.show()

IsLLP = tree.array("IsLLP")
plt.hist(IsLLP)
plt.title("Histogram of IsLLP")
plt.xlabel("0=False   1=ture")
plt.ylabel("Counts")
plt.show()
```

```python
# Plotting histograms of the decay lengths 
plt.subplot(211)
L_x = tree.array("Lx")
plt.hist(L_x, bins =50)
plt.title("Histogram of $L_x$")
plt.ylabel("Counts")
plt.xlim([0 , 4000])
plt.show()

plt.subplot(212)
L_y = tree.array("Ly")
plt.hist(L_y, bins = 50)
plt.title("Histogram of $L_y$")
plt.ylabel("Counts")
plt.xlim([0 , 4000])
plt.show()

L_z = tree.array("Lz")
plt.hist(L_z, bins = 50)
plt.title("Histogram of $L_z$")
plt.xlim([0 , 6000])
plt.ylabel("Counts")
plt.show()
```

```python
# Plotting histograms of the transverse and 3d decay lengths
L_T = np.power(L_x, 2) + np.power(L_y, 2)
L_T = np.sqrt(L_T)
plt.hist(L_T, bins = 100)
plt.title("Histogram of transverse decay length")
plt.xlim([0 , 5000])
plt.ylabel("Counts")
plt.show()


L = np.power(L_x, 2) + np.power(L_y, 2) + np.power(L_z, 2)
L = np.sqrt(L)
plt.hist(L, bins = 100)
plt.title("Histogram of 3D decay length")
plt.xlim([0 , 8000])
plt.ylabel("Counts")
plt.show()
```

```python
# Violin plot for JetPt of the all dataframes
JetPt1 = tree["JetPt"]
JetPt2 = tree2["JetPt"]
JetPt3 = tree3["JetPt"]
JetPt4 = tree4["JetPt"]
JetPt5 = tree5["JetPt"]
JetPt6 = tree6["JetPt"]

ax = sns.violinplot(data = [JetPt1, JetPt2, JetPt3, JetPt4, JetPt5, JetPt6], scale ="width")
ax.set_xticks([0, 1, 2, 3, 4, 5])
ax.set_xticklabels(['1', '2', '3', '4', '5', '6'])
ax.set_title("Violin plots for all dataframes")
ax.set_xlabel("Dataframe")
ax.set_ylabel("JetPt")
plt.show()
```

```python
# Plotting the decay lengths for each dimension for all dataframes. 
Lx1 = tree["Lx"]
Lx2 = tree2["Lx"]
Lx3 = tree3["Lx"]
Lx4 = tree4["Lx"]
Lx5 = tree5["Lx"]
Lx6 = tree6["Lx"]

ax = sns.violinplot(data = [Lx1, Lx2, Lx3, Lx4, Lx5, Lx6], scale ="width")
ax.set_xticks([0, 1, 2, 3, 4, 5])
ax.set_xticklabels(['1', '2', '3', '4', '5', '6'])
ax.set_title("x Decay lengths for all dataframes")
ax.set_xlabel("Dataframe #")
ax.set_ylabel("decay length in x")
plt.show()

Ly1 = tree["Ly"]
Ly2 = tree2["Ly"]
Ly3 = tree3["Ly"]
Ly4 = tree4["Ly"]
Ly5 = tree5["Ly"]
Ly6 = tree6["Ly"]

ax = sns.violinplot(data = [Ly1, Ly2, Ly3, Ly4, Ly5, Ly6], scale ="width")
ax.set_xticks([0, 1, 2, 3, 4, 5])
ax.set_xticklabels(['1', '2', '3', '4', '5', '6'])
ax.set_title("y Decay lengths for all dataframes")
ax.set_xlabel("Dataframe #")
ax.set_ylabel("decay length in y")
plt.show()

Lz1 = tree["Lz"]
Lz2 = tree2["Lz"]
Lz3 = tree3["Lz"]
Lz4 = tree4["Lz"]
Lz5 = tree5["Lz"]
Lz6 = tree6["Lz"]

ax = sns.violinplot(data = [Lz1, Lz2, Lz3, Lz4, Lz5, Lz6], scale ="width")
ax.set_xticks([0, 1, 2, 3, 4, 5])
ax.set_xticklabels(['1', '2', '3', '4', '5', '6'])
ax.set_title("z Decay lengths for all dataframes")
ax.set_xlabel("Dataframe #")
ax.set_ylabel("decay length in z")
plt.show()
```

```python
# Calculating the transverse and 3d decay length for each dataframe
# Plotting all of them in two violin plots

L_t1 = (Lx1*Lx1 + Ly1*Ly1)**0.5
L_t2 = (Lx2*Lx2 + Ly2*Ly2)**0.5
L_t3 = (Lx3*Lx3 + Ly3*Ly3)**0.5
L_t4 = (Lx4*Lx4 + Ly4*Ly4)**0.5
L_t5 = (Lx5*Lx5 + Ly5*Ly5)**0.5
L_t6 = (Lx6*Lx6 + Ly6*Ly6)**0.5

L_3D1 = (Lx1*Lx1 + Ly1*Ly1 + Lz1*Lz1)**0.5
L_3D2 = (Lx2*Lx2 + Ly2*Ly2 + Lz2*Lz2)**0.5
L_3D3 = (Lx3*Lx3 + Ly3*Ly3 + Ly3*Ly3)**0.5
L_3D4 = (Lx4*Lx4 + Ly4*Ly4 + Ly4*Ly4)**0.5
L_3D5 = (Lx5*Lx5 + Ly5*Ly5 + Ly5*Ly5)**0.5
L_3D6 = (Lx6*Lx6 + Ly6*Ly6 + Ly6*Ly6)**0.5


ax = sns.violinplot(data = [L_t1, L_t2, L_t3, L_t4, L_t5, L_t6], scale = 'width')
ax.set_xticks([0, 1, 2, 3, 4, 5])
ax.set_xticklabels(['1', '2', '3', '4', '5', '6'])
ax.set_title("Transverse decay lengths for all dataframes")
ax.set_xlabel("Dataframe #")
ax.set_ylabel("Transverse decay length")
plt.show()

ax = sns.violinplot(data = [L_3D1, L_3D2, L_3D3, L_3D4, L_3D5, L_3D6], scale = 'width')
ax.set_xticks([0, 1, 2, 3, 4, 5])
ax.set_xticklabels(['1', '2', '3', '4', '5', '6'])
ax.set_title("3d decay lengths for all dataframes")
ax.set_xlabel("Dataframe #")
ax.set_ylabel("3d decay length")
plt.show()
```
